# RSI Trading Strategy

* Download the data for the stock you want, by the ticker
* Set the date and the time for the time period you want to trade on
* Execute the "trades"

```python
import pandas as pd
import yfinance as yf
import numpy as np
import ta
import matplotlib.pyplot as plt

end_date = '2025-06-27'
start_date = pd.to_datetime(end_date)-pd.DateOffset(365)

stock = 'BX'

df = yf.download(tickers=stock, start=start_date, end=end_date, auto_adjust=True)

# this is just for formatting
df.columns = df.columns.get_level_values(0)
df['ticker'] = stock
df.set_index('ticker', append=True, inplace=True)
df.columns = df.columns.str.lower()

df
```

# Calculate the RSI


```python
def rsi_calc(x, window=14):
    rsi_indicator = ta.momentum.RSIIndicator(close=x, window=window)
    return rsi_indicator.rsi()

df['rsi'] = rsi_calc(df['close'], window=14)
df
```

# Plot the graph

```python
df_stock = df.xs(stock, level=1)

df_stock['rsi'].plot(label='RSI', color='blue')

plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')

plt.title('RSI')
plt.ylabel('RSI Value')
plt.legend()
plt.show()
```
